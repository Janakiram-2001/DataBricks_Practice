```python
df = spark.read.csv("/Volumes/workspace/default/volumewe47_datalake/mobile_os_usage.csv",header=True,inferSchema=True)

# Clean column names: replace spaces with underscores and remove invalid characters
for col in df.columns:
    new_col = col.replace(" ", "_") \
                 .replace(",", "") \
                 .replace(";", "") \
                 .replace("{", "") \
                 .replace("}", "") \
                 .replace("(", "") \
                 .replace(")", "") \
                 .replace("\n", "") \
                 .replace("\t", "") \
                 .replace("=", "")
    df = df.withColumnRenamed(col, new_col)

df.write.mode("overwrite").saveAsTable("default.mobile_os_usage")
display(df)
```
